```python
import numpy as np

# Datos de ejemplo (Antigüedad de la Construcción en Años y Precio de Casa en Pesos 1:100,000)
datos = np.array([
    [1, 200],
    [5, 350],
    [10, 500],
    [2, 250],
    [8, 400],
    [3, 300],
    [6, 370],
    [11, 372],
    [12, 370],
    [12, 375]
])

# Identificar Clusters con diferentes distancias
distancias = ["euclidean", "chebyshev", "manhattan"]

import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

fig = plt.figure(figsize=(12, 4))
plt.set_cmap("plasma")

for d in range(len(distancias)):
    clusters = DBSCAN(eps=2, min_samples=2, metric=distancias[d]).fit_predict(datos)
    ax = fig.add_subplot(1, 3, d+1)
    ax.scatter(datos[clusters==-1][:, 0], datos[clusters==-1][:, 1], c="cyan", s=200, marker="2")
    ax.scatter(datos[clusters>=0][:, 0], datos[clusters>=0][:, 1], c=clusters[clusters>=0], s=200, linewidth=0.5, edgecolors="black")
    ax.set_title(distancias[d].capitalize(), fontsize=24)
    ax.axis("off")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Coordenadas de dos puntos (punto A y punto B)
punto_A = np.array([1, 4])
punto_B = np.array([1, 3])

# Calcular la distancia Euclidiana entre los dos puntos
distancia_euclidiana = np.sqrt((punto_B[0] - punto_A[0])**2 + (punto_B[1] - punto_A[1])**2)

# Calcular la distancia de Chebyshev
distancia_chebyshev = np.max(np.abs(punto_B - punto_A))

# Calcular la distancia de Manhattan
distancia_manhattan = np.sum(np.abs(punto_B - punto_A))



# Crear una figura para visualizar los resultados
fig, ax = plt.subplots(figsize=(8, 6))

# Mostrar los puntos A y B en el gráfico
ax.plot(punto_A[0], punto_A[1], 'bo', label='Punto A')
ax.plot(punto_B[0], punto_B[1], 'ro', label='Punto B')

# Etiquetas de puntos A y B
ax.text(punto_A[0] + 0.2, punto_A[1], 'A', fontsize=12)
ax.text(punto_B[0] + 0.2, punto_B[1], 'B', fontsize=12)

# Mostrar las distancias calculadas
ax.text((punto_A[0] + punto_B[0]) / 2, (punto_A[1] + punto_B[1]) / 2 - 2.0, f'Euclidiana: {distancia_euclidiana:.2f}', fontsize=12, ha='center')
ax.text((punto_A[0] + punto_B[0]) / 2, (punto_A[1] + punto_B[1]) / 2 - 2.5, f'Chebyshev: {distancia_chebyshev:.2f}', fontsize=12, ha='center')
ax.text((punto_A[0] + punto_B[0]) / 2, (punto_A[1] + punto_B[1]) / 2 - 3, f'Manhattan: {distancia_manhattan:.2f}', fontsize=12, ha='center')

# Configurar el gráfico
ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
ax.set_aspect('equal')
ax.legend()

# Título
plt.title("Distancias entre dos puntos con diferentes métricas", fontsize=14)

plt.show()
```
